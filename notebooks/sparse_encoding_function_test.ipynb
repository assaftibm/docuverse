```python
import torch
def similarity(embedding1, embedding2, sp=sp):
    device = "cuda"
    embedding1 = embedding1.to(device)
    embedding2 = embedding2.to(device)
        
    emb1 = torch.zeros(sp.tokenizer.vocab_size, device=device)
    emb1[embedding1[:,0].int()] = embedding1[:,1]

    emb2 = torch.zeros(sp.tokenizer.vocab_size, device=device)
    emb2[embedding2[:,0].int()] = embedding2[:,1]

    prod =  torch.inner(emb1, emb2)
    return prod #/ (norm1*norm2) if norm1+norm2>1e-7 else 0
```

```python
similarity(r[0],r[1])/norm2(r[0])/norm2(r[1])
```

```python
def norm2(v):
    return float(torch.linalg.norm(v[:,1], ord=2))
```

```python
norm2(r[0])
```

```python
def compare(s1, s2, sp=sp):
    r=sp.encode([s1, s2])
    return similarity(r[0],r[1])/norm2(r[0])/norm2(r[1])
```

```python
compare("us", "US")
```

```python
def encode(s1, sp=sp):
    return sp.convert_token_ids_to_tokens(sp.encode([s1]))[0]
```
